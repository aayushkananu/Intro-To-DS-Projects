** Import pandas as pd.**

```python
import pandas as pd
```

** Read Salaries.csv as a dataframe called sal.**

```python
sal = pd.read_csv('Salaries.csv', low_memory = False)
sal = pd.DataFrame(sal)
```

** Check the head of the DataFrame. **

```python
sal.head()
```

** Use the .info() method to find out how many entries there are.**

```python
sal.info()
```

**What is the average BasePay ?**

```python
basepay = sal[['BasePay']] 
sal = sal[sal['BasePay'] != 'Not Provided']
average_BasePay = sal[['BasePay']].mean()
print(average_BasePay)
```

** What is the highest amount of OvertimePay in the dataset ? **

```python
sal['OvetimePay'] = sal['OvertimePay'].astype(float)
max_OvertimePay = sal['OvertimePay'].max()
print(max_OvertimePay)
```

** What is the job title of  JOSEPH DRISCOLL ? Note: Use all caps, otherwise you may get an answer that doesn't match up (there is also a lowercase Joseph Driscoll). **

```python
job_JOSEPHDRISCOLL= sal.loc[sal['EmployeeName'] == 'JOSEPH DRISCOLL']
job_JOSEPHDRISCOLL['JobTitle']
```

** How much does JOSEPH DRISCOLL make (including benefits)? **

```python
job_JOSEPHDRISCOLL['TotalPayBenefits']
```

** What is the name of highest paid person (including benefits)?**

```python
highest_TotalPay = sal[sal['TotalPayBenefits'] == sal['TotalPayBenefits'].max()]['EmployeeName']
highest_TotalPay
```

** What is the name of lowest paid person (including benefits)? Do you notice something strange about how much he or she is paid?**

```python
lowest_TotalPay = sal[sal['TotalPayBenefits']==sal['TotalPayBenefits'].min()]
lowest_TotalPay
```

** What was the average (mean) BasePay of all employees per year? (2011-2014) ? **

```python
yearly_BasePay = sal.groupby("Year")['BasePay'].mean()
yearly_BasePay
```

** How many unique job titles are there? **

```python
sal['JobTitle'].nunique()
```

** What are the top 5 most common jobs? **

```python
sal['JobTitle'].value_counts()[:5].head()
```

** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
JobTitles_in_2013 = sal[sal['Year']==2013]
sort_JobTitles_category=JobTitles_in_2013['JobTitle'].value_counts()
unique_job= sort_JobTitles_category[sort_JobTitles_category==1].count()
unique_job
```

** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
chief_in_job=0
```

```python
for job_title in sal['JobTitle'].values:
    if 'Chief' in job_title:
        chief_in_job += 1
print(chief_in_job)
```

** Bonus: Is there a correlation between length of the Job Title string and Salary? **

```python
sal['JobTitle']  = sal['JobTitle'].str.len()
```

```python
correlation = sal[['JobTitle','TotalPayBenefits']]
correlation.corr()
```
